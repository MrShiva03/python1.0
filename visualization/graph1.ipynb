```python
import pandas as pd 
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset('titanic')
```

```python
sns.countplot(df,x='survived',hue='sex', palette='RdBu')
plt.title('survived vs not survived by gender')
plt.show()
```

```python
sns.barplot(df, x='class', y='fare', hue='sex' )
plt.title('average Fare by Passengers Class and Gender')
plt.show()
```

```python
sns.violinplot(df, x='age', y='sex', hue='class')
plt.title('age distribution by Gender')
plt.show()
```

```python
sns.pointplot(df,x='class', y='survived', hue='sex', markers=['o','x'])
plt.title('Survival Rate by Class and Gender')
plt.show()
```
